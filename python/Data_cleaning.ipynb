```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# data = pd.read_csv('../sql/K6_UNEMPLOYED_RESIDENTS_BY_MARITAL_STATUS_AND_SEX.csv')
# data = pd.read_csv('../sql/K14T_RESIDENT_UNEMPLOYMENT_RATE_BY_OCCUPATION.csv')
data = pd.read_csv('../sql/K15T_UNEMPLOYED_RESIDENTS_BY_PREVIOUS_OCCUPATION_AND_SEX.csv')
# data = pd.read_csv('../sql/K16_UNEMPLOYED_RESIDENT_PMETs_AND_NON-PMETs__BY_AGE.csv')
# data = pd.read_csv('../sql/K17_LONG-TERM_UNEMPLOYED_RESIDENT_BY_PMET_STATUS_AND_AGE.csv')

data
```

```python
### ---- Step 1 ------ ###
# check for the data type and null count
data.info()

# replace null value
if (data.isnull().any().any()):
    print(data[data.isnull().any(axis=1)])
    data.fillna(0, inplace = True)
```

```python
### ---- Step 2 ------ ###
# check for Inconsistent Text and Typos
cat_col = data.select_dtypes(include=['object'])
num_col = data.select_dtypes(include=['int','float64'])
[data[cat].unique() for cat in cat_col]
```

```python
### ---- Step 3 ------ ###
# Find duplicate rows
data[data.duplicated(keep=False)]
data.drop_duplicates(inplace=True)
data
```

```python
### ---- Step 4 ------ ###
# Find outliers
data.describe()
```

```python
data.hist();
```

```python
# longitudinal 
num_col_label = num_col.copy()
num_col_label['label'] = cat_col.apply(lambda x: '_'.join(x.astype(str)), axis=1)
num_col_label_T = num_col_label.T.rename(columns=num_col_label.T.loc['label']).drop('label').apply(pd.to_numeric, errors='coerce')
num_col_label_T.hist();
```

```python
# box plot
print(num_col)
for col in num_col.columns:
    plt.title(f'Curve for {col}')
    sns.boxplot(y=num_col[col])
    plt.show();
```

```python
# box plot - longitudinal 
#print(num_col_label_T)
for col in num_col_label_T.columns:
    plt.title(f'Curve for {col}')
    sns.boxplot(y=num_col_label_T[col])
    plt.show();
```

```python
# outlier using the quartiles - cross-sectional
for col in num_col.columns:
    q25, q50, q75 = np.percentile(num_col_label[col], [25, 50, 75])
    iqr = q75 - q25
    
    # calculate the min and max
    min_bound = q25 - 1.5*iqr
    max_bound = q75 + 1.5*iqr
    
    # print(min_bound, q25, q50, q75, max_bound)
    outlier = num_col_label[[col,'label']][(num_col_label[col] < min_bound ) | (num_col_label[col] > max_bound )]
    if len(outlier)>0:
        print(outlier)

# Extract whole row
q1 = num_col.quantile(0.25)
q3 = num_col.quantile(0.75)
iqr = q3 - q1
is_outlier = (num_col < (q1 - 1.5*iqr) ) | (num_col > (q3 + 1.5*iqr) )
rows_with_outliers = is_outlier.any(axis=1)
outlier_rows = data[rows_with_outliers]
outlier_rows
```

```python
# outlier using the quartiles - longitudinal 
#print(num_col_label_T)
for col in num_col_label_T.columns:
    q25, q50, q75 = np.percentile(num_col_label_T[col], [25, 50, 75])
    iqr = q75 - q25
    
    # calculate the min and max
    min_bound = q25 - 1.5*iqr
    max_bound = q75 + 1.5*iqr
    
    #print(min_bound, q25, q50, q75, max_bound)
    outlier = num_col_label_T[[col]][(num_col_label_T[col] < min_bound ) | (num_col_label_T[col] > max_bound )]
    if len(outlier)>0:
        print(outlier)

# Extract whole row
q1 = num_col_label_T.quantile(0.25)
q3 = num_col_label_T.quantile(0.75)
iqr = q3 - q1
is_outlier = (num_col_label_T < (q1 - 1.5*iqr) ) | (num_col_label_T > (q3 + 1.5*iqr) )
rows_with_outliers = is_outlier.any(axis=1)
outlier_rows = num_col_label_T[rows_with_outliers]
outlier_rows
```
