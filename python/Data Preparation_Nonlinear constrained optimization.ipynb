```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize

# ------------------------------
# Optimization function
# ------------------------------
def huber_loss(error, delta=0.1):
    return np.where(np.abs(error) <= delta, 0.5 * error**2, delta * (np.abs(error) - 0.5 * delta))

def objective(w, array1, array2):
    W = w.reshape(len(array1), len(array2))
    return np.sum((np.dot(W, array2) - array1) ** 2)

    # errors = np.dot(W, array2) - array1
    # return np.sum(errors ** 4)  # Higher-order penalty
    
    # errors = np.dot(W, array2) - array1
    # max_error_penalty = 10 * np.max(np.abs(errors))  # Scale this as needed
    # return np.sum(errors**2) + max_error_penalty

    # max_allowed_error=0.1
    # errors = np.abs(np.dot(W, array2) - array1)
    # return max_allowed_error - np.max(errors)
    # errors = np.dot(W, array2) - array1
    # return np.sum(huber_loss(errors))


def optimize_weights(array1, array1_labels, array2, array2_labels):
    array1 = np.asarray(array1)
    array2 = np.asarray(array2)

    # Constraint: each column (age group) sums to 1
    constraints = [
        {'type': 'eq', 'fun': lambda w, idx=idx: np.sum(w.reshape(len(array1), len(array2))[:, idx]) - 1}
        for idx in range(len(array2))
    ]
    
    w0 = np.full(len(array1) * len(array2), 1.0 / len(array1))
    bounds = [(0, 1)] * len(w0)

    result = minimize(objective, w0, args=(array1, array2), method='SLSQP', constraints=constraints, bounds=bounds)
    weights_matrix = result.x.reshape(len(array1), len(array2))

    weights_df = pd.DataFrame(
        weights_matrix,
        index=array1_labels.tolist(),
        columns=array2_labels.tolist()
    )

    weights_df['Original k15T'] = array1
    weights_df['Calculated k15T'] = weights_df[array2_labels.tolist()].dot(array2)
    weights_df['Error'] = weights_df['Calculated k15T'] - weights_df['Original k15T']
    
    return weights_df

def process_split (year, Pmet_Status):                                                                                                                    # ------------------------------
    # Apply optimization
    # ------------------------------
    k15_vals = K15T_group_Pmet[year][K15T_group_Pmet['Pmet_Status'] == Pmet_Status]
    k15_labels = K15T_group_Pmet['Gender'][K15T_group_Pmet['Pmet_Status'] == Pmet_Status]
    k16_vals = K16_drop[year][K16_drop['Pmet_Status'] == Pmet_Status]
    k16_labels = K16_drop['Age'][K16_drop['Pmet_Status'] == Pmet_Status]
    
    weights_df = optimize_weights(k15_vals, k15_labels, k16_vals, k16_labels)
    
    #print("Detailed Results Table:")
    #print("-----------------------")
    #print(weights_df)
    
    # ------------------------------
    # Melt weights table to long format
    # ------------------------------
    melted = weights_df.reset_index(names='Gender').melt(
        id_vars='Gender',
        value_vars=k16_labels.tolist(),
        var_name='Age_Group',
        value_name='Weight'
    )
    #print(melted)
    
    # ------------------------------
    # Merge to apply weights (no loop needed)
    # ------------------------------
    k16_values = K16_drop[['Age', year]][K16_drop['Pmet_Status'] == Pmet_Status].drop_duplicates()
    melted = melted.merge(k16_values, left_on='Age_Group', right_on='Age', how='left')
    melted['Count'] = melted['Weight'] * melted[year]
    melted = melted.drop(columns=['Age', year])
    melted = melted.reset_index(drop=True)
    
    # Final result
    melted['Year'] = year
    return melted

# Initialize empty list to store results
all_results = []

# Get unique Pmet_Status values from the combined dataset (you can use either K15T_group_Pmet or K16_drop)
pmet_status_list = K15T_group_Pmet['Pmet_Status'].unique()

# Loop through each year and Pmet_Status
for year in range(2014, 2025):
    year_str = str(year)
    for pmet_status in pmet_status_list:
        try:
            melted = process_split(year_str, pmet_status)
            melted['Pmet_Status'] = pmet_status  # Add status as a column
            all_results.append(melted)
        except KeyError:
            print(f"Year {year} or status '{pmet_status}' not found in data — skipping.")
        except Exception as e:
            print(f"Error processing year {year} and status '{pmet_status}': {e}")

# Combine all results into one DataFrame
final_result = pd.concat(all_results, ignore_index=True)

# View result
print(final_result)
```

```python
def get_occupation_split(year, gender, pmet_status):
    year_str = str(year)

    # Get K15 values and labels (Occupation totals)
    k15_vals = K15T_drop [(K15T_drop ['Pmet_Status'] == pmet_status) & (K15T_drop ['Gender'] == gender)][year_str]
    k15_labels = K15T_drop [(K15T_drop ['Pmet_Status'] == pmet_status) & (K15T_drop ['Gender'] == gender)]['Occupation']

    # Get K16 values and labels (Age group totals from final_result)
    k16_filtered = final_result[
        (final_result['Year'] == year_str) &
        (final_result['Gender'] == gender) &
        (final_result['Pmet_Status'] == pmet_status)
    ]
    k16_vals = k16_filtered['Count']
    k16_labels = k16_filtered['Age_Group']

    # Run optimization
    weights_df = optimize_weights(k15_vals, k15_labels, k16_vals, k16_labels)

    # Melt to long format
    melted = weights_df.reset_index(names='Occupation').melt(
        id_vars='Occupation',
        value_vars=k16_labels.tolist(),
        var_name='Age_Group',
        value_name='Weight'
    )

    # Add columns for merging
    melted['Year'] = year_str
    melted['Gender'] = gender
    melted['Pmet_Status'] = pmet_status

    # Merge with final_result to get age group marginal totals
    merge_keys = ['Age_Group', 'Year', 'Gender', 'Pmet_Status']
    melted = melted.merge(
        final_result[merge_keys + ['Count']],
        on=merge_keys,
        how='left'
    )

    # Compute weighted counts per occupation per age group
    melted['Occupation_Count'] = melted['Weight'] * melted['Count']

    # Clean up
    melted = melted[['Occupation', 'Age_Group', 'Weight', 'Occupation_Count', 'Year', 'Gender', 'Pmet_Status']]
    
    return melted

get_occupation_split(2014, 'Female', 'Non-PMETs')
```

```python
def get_split(year, age, pmet_status):
    year_str = str(year)

    # Get K15 values and labels (Occupation totals)
    k17_vals = K17[(K17['Age'] == age ) & (K17['Pmet_Status'] == pmet_status)][year_str] 
    k17_labels = K17[(K17['Age'] == age ) & (K17['Pmet_Status'] == pmet_status)]['Status']
    # print(k17_vals)
    # print(k17_labels)
    
    # Get K16 values and labels (Age group totals from final_result)
    occupation_split_df_grouped = (occupation_split_df
                                    .groupby(['Pmet_Status','Age_Group','Year','Occupation'], as_index=False)
                                    .sum()
                                    .drop(['Gender'], axis =1))
    occupation_split_df_filtered = (occupation_split_df_grouped
                                    .loc[lambda df: (df['Year'] == year_str ) & (df['Age_Group'] == age) & (df['Pmet_Status'] == pmet_status)])
    occupation_split_grouped_vals = occupation_split_df_filtered['Occupation_Count']
    occupation_split_grouped_labels = occupation_split_df_filtered['Occupation']
    # print(occupation_split_grouped_vals)
    # print(occupation_split_grouped_labels)
    
    # Run optimization
    weights_df = optimize_weights(occupation_split_grouped_vals, occupation_split_grouped_labels,k17_vals, k17_labels)
    # print(weights_df)
    
    # Melt to long format
    melted = weights_df.reset_index(names='Occupation').melt(
        id_vars='Occupation',
        value_vars=k17_labels.tolist(),
        var_name='Status',
        value_name='Weight'
    )
    
    # Add columns for merging
    melted['Year'] = year_str
    melted['Age_Group'] = age
    melted['Pmet_Status'] = pmet_status
    
    print(melted)
    # Merge with final_result to get age group marginal totals
    merge_keys = ['Pmet_Status', 'Age_Group', 'Status']
    table_K17 = K17[(K17['Age'] == age ) & (K17['Pmet_Status'] == pmet_status)].rename(columns={year_str: 'Occupation_Count', 'Age': 'Age_Group'}, inplace=False)
    #print(table_K17)
    melted = melted.merge(
        table_K17[merge_keys + ['Occupation_Count']],
        on=merge_keys,
        how='left'
    )
    
    melted['Status_Count'] = melted['Weight'] * melted['Occupation_Count']  

    # Clean up
    melted = melted[['Occupation', 'Age_Group', 'Status_Count', 'Year', 'Pmet_Status', 'Status']]

    return melted

melted = get_split(2014, '15 - 29', 'PMETs')
melted
```
